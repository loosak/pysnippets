```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize': (16, 8),})
```

```python
# Definujeme si vstupní data, v tomto případě 1 a 3
# Pokud je vstup 1, chceme výstup 3
x = np.array([[1], [3]])
# Pokud je vstup 3, chceme výstup 5
y = np.array([[3], [5]])
```

```python
# Vytvoříme regressor neboli lineární funkci
reg = LinearRegression()
# Optimalizujeme
reg.fit(x, y)
```

```python
reg.score(x, y)
```

```python
# Pomocí funkce plot() se vykreslí čárový graf
plt.plot(x, reg.predict(x))
# Pomocí funkce scatter() vykreslíme jednotlivé body v grafu
# V tomto případě body [1, 3] a [3, 5]
plt.scatter(x, y, c="r") # v parametru c specifikujeme styl bodů, chceme je červené
```

```python
a = reg.coef_
b = reg.intercept_
f"Sklon je {a} a bias je {b}"
```

```python
import requests
r = requests.get('http://sipi.usc.edu/database/download.php?vol=misc&img=4.2.03', allow_redirects=True)
open('Mandrill.tiff', 'wb').write(r.content)
r.status_code
```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

mandrill = Image.open('Mandrill.tiff')
plt.imshow(np.asarray(mandrill))
plt.title('Mandrill')
plt.axis('off')
```

```python
mandrill = np.asarray(mandrill)
hist, bins = np.histogram(mandrill, bins=128, range=(0, 255))
plt.bar(bins[:-1], hist)
```

```python
mandrill = np.asarray(mandrill)
b = mandrill[:, :, 0]
g = mandrill[:, :, 1]
r = mandrill[:, :, 2]
plt.hist(r.ravel(), bins=256, range=(0, 255), color='r')
plt.hist(g.ravel(), bins=256, range=(0, 255), color='g')
plt.hist(b.ravel(), bins=256, range=(0, 255), color='b')
plt.title('Histogram')
```

```python
hist, bins = np.histogram(mandrill[:, :, 0], bins=128, range=(0, 255))
plt.plot(hist, color='b')
hist, bins = np.histogram(mandrill[:, :, 1], bins=128, range=(0, 255))
plt.plot(hist, color='g')
hist, bins = np.histogram(mandrill[:, :, 2], bins=128, range=(0, 255))
plt.plot(hist, color='r')
```
