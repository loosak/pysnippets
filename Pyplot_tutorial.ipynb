# Shows how to use latex in a plot

```python
#plt.rc('text', usetex=True)

# interface tracking profiles
N = 500
delta = 0.6
x = np.linspace(-1, 1, N)
plt.plot(x, (1 -   np.tanh(4*x/delta)) / 2,          # phase field tanh profiles
         x, (1.4 + np.tanh(4*x/delta)) / 4, "C2",    # composition profile
         x, x < 0, 'k--',                             # sharp interface
        ) 

# x axis
plt.xticks((-1, 0, 1), ('$-1$', r'$\pm 0$', '$+1$'), color='k', size=20)


# Left Y-axis labels, combine math mode and text mode
plt.ylabel(r'\bf{phase field} $\phi$', {'color': 'C0', 'fontsize': 20})
plt.yticks((0, 0.5, 1), (0, .5, 1), color='k', size=20)

# Right Y-axis labels
plt.text(1.02, 0.5, r"\bf{level set} $\phi$", {'color': 'C2', 'fontsize': 20},
         horizontalalignment='left',
         verticalalignment='center',
         rotation=90,
         clip_on=False,
         transform=plt.gca().transAxes)

# legend
plt.legend(('phase field', 'level set', 'sharp interface'),
           shadow=True, loc=(0.01, 0.48), handlelength=1.5, fontsize=16)

# the arrow
plt.annotate("", xy=(-delta / 2., 0.1), xytext=(delta / 2., 0.1),
             arrowprops=dict(arrowstyle="<->", connectionstyle="arc3"))
plt.text(0, 0.1, r'$\delta$',
         {'color': 'black', 'fontsize': 24, 'ha': 'center', 'va': 'center',
          'bbox': dict(boxstyle="round", fc="white", ec="black", pad=0.2)})
```

```python
fig, ax = plt.subplots()
ax.tick_params(width=1)
ax.plot(range(10))
```

```python
import numpy as np
fig, ax = plt.subplots()

t = np.arange(0.0, 100.0, 0.01)
s = np.sin(2 * np.pi * t) * np.exp(-t * 0.01)

ax.plot(t, s)
```

```python
import matplotlib.ticker

fig, ax = plt.subplots()

ax.xaxis.set_minor_locator(matplotlib.ticker.AutoMinorLocator())
ax.tick_params(which='both',  width=2)
ax.tick_params(which='major', length=7)
#ax.tick_params(which='minor', length=4, color='r')
ax.plot(t, s)
```

```python
x = np.linspace(-2*np.pi,2*np.pi)

fig, axs = plt.subplots(2)

axs[0].plot(x, np.sin(x), color='g', label=r'$f(x) = \sin(x)$')
axs[0].plot(x, np.cos(x), label=r'$f(x) = \cos(x)$')
axs[1].plot(x, np.cos(2 * np.pi * x) * np.exp(-x), label=r'$f(x) = \cos(2 \pi x) \exp(-x)$')
```

# Plots with different scales


```python
x = np.linspace(-2*np.pi,2*np.pi,90)

fig, ax1 = plt.subplots()
color = 'tab:red'
ax1.plot(x, np.exp(x), color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_ylabel(r'$\exp(x)$', color=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.plot(x, np.sin(x), color=color)
ax2.tick_params(axis='y', labelcolor=color)
fig.tight_layout()  # otherwise the right y-label is slightly clipped
```

```python
fig, ax = plt.subplots(2, sharex=True)

color = 'tab:red'
ax[0].plot(x, np.exp(x), color=color)
ax[0].tick_params(axis='y', labelcolor=color)
ax[0].set_ylabel(r'$\exp(x)$', color=color)

color = 'tab:blue'
ax[1].plot(x, np.sin(x), color=color)
ax[1].tick_params(axis='y', labelcolor=color)
```
