```python
import pandas as pd
import matplotlib.pyplot as plt
import watermark
import requests

plt.style.use('./d4sci.mplstyle')
%matplotlib inline
%load_ext watermark
%watermark -n -v -m -g -iv
```

```python
r = requests.get('https://www.worldometers.info/coronavirus/')
corona = pd.read_html(r.text)[0]
# corona.drop(['World','Total:'], inplace=True)
corona_total = corona.iloc[:1]
corona = corona.iloc[1:-1]
corona.rename(columns={
                    'Country,Other': 'Country',
              },
              inplace=True,
)

# Population by country
r = requests.get('https://www.worldometers.info/world-population/population-by-country/')
countr = pd.read_html(r.text)[0]
countr.rename(
            columns={
                     'Country (or dependency)': 'Country',
                     'Population (2020)': 'Population',
            },
            inplace=True,
)
# countr.replace(
countr['Country'].replace(
            to_replace={
                     'Central African Republic': 'CAR',
                     'Czech Republic (Czechia)': 'Czechia',
                     "Côte d'Ivoire": 'Ivory Coast',
                     'Saint Barthelemy': 'St. Barth',
                     'Saint Kitts & Nevis': 'Saint Kitts and Nevis',
                     'Saint Pierre & Miquelon': 'Saint Pierre Miquelon',
                     'South Korea': 'S. Korea',
                     'St. Vincent & Grenadines': 'St. Vincent Grenadines',
                     'State of Palestine': 'Palestine',
                     'United Arab Emirates': 'UAE',
                     'United Kingdom': 'UK',
                     'United States': 'USA',
            },
            inplace=True,
)

df = pd.merge(corona,
         countr[['Country', 'Population', 'Density (P/Km²)', 'Land Area (Km²)', 'Migrants (net)', 'Med. Age', 'Urban Pop %']],
         on='Country')

df.set_index('Country', inplace=True)

del corona
del countr

df.sort_values('TotalDeaths', ascending=False)
```
