````python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize': (14, 10),})
````

````python
df = pd.read_csv('alerts.2020.08.csv', sep=',')
df['Time'] = pd.to_datetime(df['Occurrence Time'])
df.drop(columns=['Occurrence Time','Alert Name', 'Severity', 'Alert Source'], inplace=True)
df['Day'] = df['Time'].apply(lambda x: pd.Timestamp.to_pydatetime(x).strftime("%A"))
df.Violation = df.Violation.str.strip(' ms/op')
df.drop(df[df.Violation == 'Warning'].index.values.tolist(),inplace=True)
df.Violation = pd.to_numeric(df.Violation)
#df.Violation = df.Violation.apply(lambda x: float(x))

#df = df[(df['Time'] > dt.datetime(2020,7,1))] # last month
df.set_index('Time', inplace=True)
df =  df[df.index.month > dt.datetime.now().month-1] 

df
````
