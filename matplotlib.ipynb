```python
import sys
import numpy as np

theta1 = np.linspace(-2*np.pi,2*np.pi,360)
sys.getsizeof(theta1)
print(*theta1[:20])
type(theta1),sys.getsizeof(theta1[:20])
```

```python
delta = 4*np.pi/360
delta
```

```python
theta2 = [x*delta for x in range(-180,181)]
sys.getsizeof(theta2)
```

```python
len(theta2),theta2[0],theta2[-1]
```

```python
print(*theta2[:20])
sys.getsizeof(theta2[:20])
```

```python
theta3 = np.asarray(theta2)
sys.getsizeof(theta3)
```
